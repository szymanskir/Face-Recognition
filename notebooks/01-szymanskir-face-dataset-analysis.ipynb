The goal of this notebook is to analyse the samples of the olivetti faces dataset whether any image preprocessing is required before proceeding to the feature extraction stage.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import fetch_olivetti_faces
from skimage import exposure
```

```python
dataset = fetch_olivetti_faces()
```

```python
fig, axes = plt.subplots(1,2)
axes[0].imshow(dataset.images[398], cmap='gray')
axes[1].imshow(dataset.images[399], cmap='gray')
```

Samples were taken using different angles, as a result some samples have different lightining. In order to make all samples equally illuminated histogram equalisation will be applied.

```python
fig, axes = plt.subplots(1,2)
axes[0].imshow(exposure.equalize_hist(dataset.images[398]), cmap='gray')
axes[1].imshow(exposure.equalize_hist(dataset.images[399]), cmap='gray')
```
