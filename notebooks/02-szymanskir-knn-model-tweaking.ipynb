The goal of this notebook is to choose the `n_neighbors` coefficient for the KNN model.

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from src.models.knn_model import create_knn_model
```

```python
features_train_filepath = os.path.join('..', 'data', 'processed', 'features_train.csv')
features_test_filepath = os.path.join('..', 'data', 'processed', 'features_test.csv')
labels_train_filepath = os.path.join('..', 'data', 'raw', 'labels_train.csv')
labels_test_filepath = os.path.join('..', 'data', 'raw', 'labels_test.csv')
```

```python
features_train = pd.read_csv(features_train_filepath)
features_test = pd.read_csv(features_test_filepath)
labels_train = pd.read_csv(labels_train_filepath)
labels_test = pd.read_csv(labels_test_filepath)
```

```python
scores = list()
for coefficient in range(1, len(features_train)):
    model = create_knn_model(coefficient, features_train, labels_train.values.ravel())
    scores.append(model.score(features_test, labels_test))
```

```python
fig = plt.figure(figsize=(16, 12))
plot = sns.scatterplot(x=range(1, len(features_train)), y=scores)
plot.set(xlabel='Number of nearest neighbors', ylabel='Accuracy score of the prediction')
```

The 1NN model had the best performance.
